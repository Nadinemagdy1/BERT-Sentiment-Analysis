```python
from transformers import pipeline

print("Loading model...")

sentiment = pipeline("sentiment-analysis")

print("Model loaded!")

reviews = [
    "This restaurant is amazing, the food is delicious!",
    "The service was very slow and the food was cold.",
    "The place is nice but the prices are too high."
]

for review in reviews:
    result = sentiment(review)
    print(review)
    print(result)
    print()
```
